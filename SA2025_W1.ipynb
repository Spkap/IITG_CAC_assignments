### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
#your code here
med = cars['mpg'].median()
above = cars[cars['mpg'] > med]
above['ratio'] = above['horsepower'] / above['weight']
max_r = above['ratio'].max()
best = above[above['ratio'] == max_r]
print(best)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg = cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10, 6))

for origin in avg_mpg['origin'].unique():
    data = avg_mpg[avg_mpg['origin'] == origin]
    plt.plot(data['model_year'], data['mpg'], marker='o', label=origin.title())

plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.title('Evolution of Average MPG by Origin')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reset index to avoid duplicate labels error
data = cars.reset_index()

plt.figure(figsize=(10, 7))

sns.scatterplot(data=data,
                x='horsepower',
                y='weight',
                hue='origin',
                size='mpg',
                hue_order=['japan', 'europe', 'usa'],
                alpha=0.7)

plt.title('Car Weight vs Horsepower by Origin and Fuel Efficiency')
plt.xlabel('Horsepower')
plt.ylabel('Weight (lbs)')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
cars = pd.read_csv('/Users/sourabhkapure/Downloads/Cars.csv')

car_stats = cars.groupby('name').agg({
    'mpg': ['count', 'mean', 'std'],
    'model_year': 'nunique'
}).round(3)


car_stats.columns = ['count', 'avg_mpg', 'mpg_std', 'years']

consistent = car_stats[
    (car_stats['years'] > 1) &
    (car_stats['count'] >= 2) &
    (car_stats['mpg_std'] < 1.0)
].copy()

consistent = consistent.sort_values(['count', 'avg_mpg'], ascending=[False, False])

print("Consistent car models:")
print(consistent)
```
